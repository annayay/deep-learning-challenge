```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
import pandas as pd
```

```python
# Import pandas and read the charity_data.csv from the provided cloud URL.

application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN', 'NAME'])
```

```python
application_df.head()
```

```python
# Determine the number of unique values in each column.
unique_values = application_df.nunique()
print(unique_values)
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()

# Step 2: Identify values that occur 17 times or fewer
to_replace = application_type_counts[application_type_counts <= 17].index

# Step 3: Replace those values with "Other"
application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(to_replace, "Other")

# Display the updated value counts to confirm changes
updated_application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
print(updated_application_type_counts)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts = application_df['CLASSIFICATION'].value_counts()

# Step 2: Display the counts to identify which values to replace
print(classification_counts)

# Step 3: Identify values to replace with "Other"
to_replace = classification_counts[classification_counts <= 71].index

# Step 4: Replace those values with "Other"
application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(to_replace, "Other")

# Display the updated value counts to confirm changes
updated_classification_counts = application_df['CLASSIFICATION'].value_counts()
print(updated_classification_counts)
```

```python
# Step 1: Set a cutoff value
cutoff_value = 6 

# Step 2: Get the value counts for CLASSIFICATION
classification_counts = application_df['CLASSIFICATION'].value_counts()

# Step 3: Create a list of classifications to be replaced
classifications_to_replace = classification_counts[classification_counts < cutoff_value].index.tolist()

# Step 4: Replace in the DataFrame
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls, "Other")

# Step 5: Check to make sure the replacement was successful
updated_classification_counts = application_df['CLASSIFICATION'].value_counts()
print(updated_classification_counts)
```

```python
# Convert categorical data to numeric using pd.get_dummies
application_df_encoded = pd.get_dummies(application_df, drop_first=True)

# Display the first few rows of the new DataFrame to confirm the conversion
print(application_df_encoded.head())
```

```python
# Define the target variable
y = application_df_encoded['IS_SUCCESSFUL']  
# Step 2: Define the features
X = application_df_encoded.drop(columns=['IS_SUCCESSFUL'])  

# Step 3: Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)  # 80% training, 20% testing

# Display the shapes of the resulting datasets to confirm the split
print(f"Training feature set shape: {X_train.shape}")
print(f"Testing feature set shape: {X_test.shape}")
print(f"Training target set shape: {y_train.shape}")
print(f"Testing target set shape: {y_test.shape}")
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```
